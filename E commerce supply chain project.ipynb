```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
df = pd.read_excel(r"/python day 29 supply_chain_data.xlsx")
df
```

```python
df.describe()
```

```python
figure = px.scatter(df,x='Price',y='Revenue generated',color='Product type')
figure.update_layout(title="product type wise revenue generated")
figure.show()
```

```python
df_grouped = df.groupby('Product type')['Number of products sold'].sum().reset_index()
df_grouped
```

```python
a = px.pie(df_grouped,values = 'Number of products sold',names = 'Product type',hole = 0.4,color_discrete_sequence=px.colors.qualitative.Pastel)
a.update_traces(textposition='inside',textinfo='percent+label')
a.show()
```

```python
total_revenue=df.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
total_revenue
```
